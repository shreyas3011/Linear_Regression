**Simple Linear Regression:**

```python
#'Simple linear regression involves only one independent variable and one dependent variable.The relationship between the independent and dependent
#variables is assumed to be linear and the model aims to find the best-fitting straight line that minimizes the sum of squared residuals'.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('placement2.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
sns.scatterplot(x='cgpa',y='package',data=df)
```

```python
x=df.iloc[:,0:1]
y=df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
x_test.head()
```

```python
x_test.tail()
```

```python
#y_pred=mx+b
```

```python
m=lr.coef_
```

```python
m
```

```python
b=lr.intercept_
```

```python
b
```

```python
y_pred=m*6.37+b
y_pred
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

**best fit line**

```python

plt.scatter(df['cgpa'],df['package'])
plt.plot(x_train,lr.predict(x_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')

```
